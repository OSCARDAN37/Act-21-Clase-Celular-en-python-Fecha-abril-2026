```python
class Celular:
    # Constructor con atributos
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")

    def mostrar_info(self):
        print(f"El celular {self.marca} {self.modelo}, Precio: ${self.precio}")

celular1 = Celular("Apple", "iPhone 15 Pro", 1200)
celular2 = Celular("Xiaomi", "Redmi Note 12", 300)

print(" Celular 1:")
celular1.mostrar_info()
celular1.encender()

print("Celular 2:")
celular2.mostrar_info()
celular2.encender()

print("Programa terminado")
print("Creado de software pro Oscar Flores NL = 19")
```
